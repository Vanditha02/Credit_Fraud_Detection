# Under Sampling

```python
normal_sample=normal.sample(n=492)
```

```python
normal_sample.head()
```

```python
df1=pd.concat([normal_sample,fr],axis=0)
df1.head()
```

```python
df1.tail()
```

```python
df1['Class'].value_counts()
```

```python
df1.groupby('Class').mean()
```

```python
X=df1.drop(columns='Class',axis=1)
Y=df1['Class']
```

```python
X
```

```python
Y
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
X_train_pred=model.predict(X_train)
Tr_accuracy=accuracy_score(X_train_pred,Y_train)
```

```python
Tr_accuracy
```

```python
X_test_pred=model.predict(X_test)
Test_Accuracy=accuracy_score(X_test_pred,Y_test)
Test_Accuracy
```

```python
from sklearn.neighbors import KNeighborsClassifier
knnmodel=KNeighborsClassifier(n_neighbors=2)
knnmodel.fit(X_train,Y_train)
Y_pred = knnmodel.predict(X_test)
Y_train_pred=knnmodel.predict(X_train)
```

```python
Tr_accuracy=accuracy_score(Y_train_pred,Y_train)
Tr_accuracy
```
